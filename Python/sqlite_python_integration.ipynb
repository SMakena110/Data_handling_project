### Working With SQLITE and Python

###### Let us connect to our sql database

##### 1. Import reqired library
We will need:

```python
import pandas as pd
import sqlite3
```

##### 2. Connect to existing SQLITE database

```python
connection = sqlite3.connect("C:\\Users\\Hp\\Downloads\\hr")
cursor = connection.cursor()
```

##### 3. Create a Cursor object

```python
cursor = connection.cursor()
```

##### 4. View all existing Tables in the database

```python
# setup sql statement
sql = '''
         SELECT name AS table_name
         FROM sqlite_master 
         WHERE type = 'table';
      '''
# execute the sql statement
cursor.execute(sql)
```

```python
# print records from sql execution
rows = cursor.fetchall()
for row in rows:
    print(row[0].upper())
```

```python
# fetch all the records in the Country table
country_sql = '''
                 SELECT * 
                 FROM COUNTRIES
                 LIMIT 100;
              '''
cursor.execute(country_sql)

country_records = cursor.fetchall()

for row in country_records:
    print(row)
```

```python
# convert the output rows into dataframe
country_df = pd.DataFrame(country_records[1:], columns = country_records[0])
country_df
```

```python
country_df.columns = ['Country ID', 'Country Name', 'Region ID']
country_df
```

#### 5. Using Pandas Read_sql method

```python
df = pd.read_sql(country_sql, connection)
df.rename(columns = {"country_id":"Country ID","country_name": "Country Name", "region_id":"Region ID"}, inplace = True)
df.head(5)
```
