
#### REPORT ON THE ESTIMATION OF OBESITY LEVELS BASED ON DIETARY AND PHYSICAL HABITS


1. **Data Preparation** :The dataset contained 2,111 records across 17 features, including demographic, physical, dietary, and activity data. Key steps in preparing the data were: 
- **Completeness**: All fields had non-null values.
- **Data Cleaning**: Duplicate entries (42 records) were removed, leaving 2,069 clean records. Inconsistent column naming conventions were standardized. 
- **Handling Outliers**: Weight outliers were addressed through capping between the 15th and 85th percentiles, normalizing weight distribution. 
- **Normalization:** Continuous variables (age, weight, height, physical activity frequency, etc.) were normalized using Min-Max Scaling to facilitate model performance and feature comparability. Encoding: Categorical variables were encoded into numerical formats using LabelEncoder for compatibility with machine learning algorithms.
