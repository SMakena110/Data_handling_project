### 3.How to rename column headings

#### 6. How to filter records that are relevant to our analysis/ by specific columns

```python
#show me only records where country is Kenya.
df_kenya = df[df["country"] =="Kenya"]
df_kenya.head(5)
```

```python
df_uganda = df[df["country"] == "Uganda"]
df_uganda.drop(["country", "iso_code"], axis=1, inplace=True)
df_uganda.head()
```

#### 7. How to sort the data by a specific column

```python
# sort the data by new_deaths column in a descending manner
df_kenya.sort_values(by=["new_deaths", "date_vaccinations"], inplace=True, ascending=(1,0))
df_kenya
```

#### 8. Is the data Consistent

```python
df_kenya_sorted = df_kenya.sort_values(by=["date_vaccinations"], ascending= True)
df_kenya_sorted
```

```python
# This is similar to calculated fields in Tableau
```

```python
# get the ratio of people vaccinated against population
#vaccinated_ratio = population/people_fully_vaccinated
df_kenya_sorted["vaccinated_ratio"] = df_kenya_sorted["people_fully_vaccinated"]/df_kenya_sorted["population"]
df_kenya_sorted.tail()
```
