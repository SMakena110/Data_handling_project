```python
import pandas as pd
import statsmodels as sm
import statistics as stats
import numpy as np
```

```python
Public_debt = pd.read_csv("data/Public Debt (Ksh Million).csv")
Public_debt.head()
```

```python
Public_debt.info()
```

```python
# Convert Domestic Debt, External Debt to a float value

Public_debt["Domestic Debt"] = (Public_debt["Domestic Debt"]).apply(lambda x : x.replace(",",""))
Public_debt["Domestic Debt"] = (Public_debt["Domestic Debt"]).apply(lambda x : float(x))

Public_debt["External Debt"] = (Public_debt["External Debt"]).apply(lambda x : x.replace(",",""))
Public_debt["External Debt"] = (Public_debt["External Debt"]).apply(lambda x : float(x))

Public_debt["Total"] = (Public_debt["Total"]).apply(lambda x : x.replace(",",""))
Public_debt["Total"] = (Public_debt["Total"]).apply(lambda x : float(x))
Public_debt
```

```python
Public_debt.describe()
```

#### Introduce Standard Statistics

```python
# Max
print(np.max(Public_debt["Domestic Debt"]))

# Min
print(np.min(Public_debt["Domestic Debt"]))

# Standard Deviation
print(np.std(Public_debt["Domestic Debt"]))

# mode
print(stats.mode(Public_debt["Domestic Debt"]))

#median
print(np.median(Public_debt["Domestic Debt"]))

# Count
# print(np.count_nonzero(Public_debt["Domestic Debt"]))
print(len(Public_debt["Domestic Debt"]))
# Percentiles
# np.percentile(Public_debt["Domestic Debt"])

# Quartiles
print(stats.quantiles(Public_debt["Domestic Debt"]))
```
